```python
genre_profit = {}
for genre in genre_cols:
    genre_profit[genre] = data_budget_revenue[data_budget_revenue[genre] == 1]['return_on_invest'].mean()

genre_profit_df = pd.DataFrame(genre_profit.items(), columns=['Genre', 'Average ROI'])
genre_profit_df.sort_values('Average ROI', ascending=False, inplace=True)
plt.figure(figsize=(12,6))
sns.barplot(data=genre_profit_df, x='Genre', y='Average ROI', hue='Genre', palette='viridis', legend=False)
plt.xticks(rotation=45)
plt.title('Average ROI by Genre')
plt.show()
```

```python
yearly = data_budget_revenue.groupby('release_year')[['budget','revenue']].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly, x='release_year', y='budget', label='Avg Budget')
sns.lineplot(data=yearly, x='release_year', y='revenue', label='Avg Revenue')
plt.title('Average Budget and Revenue Over Time')
plt.legend()
plt.show()
```
